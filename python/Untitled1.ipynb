```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data points from a normal distribution
mu = 0  # mean
sigma = 1  # standard deviation
data_points = np.random.normal(mu, sigma, 1000)

# Plot the histogram of the data
plt.hist(data_points, bins=50, density=True, alpha=0.6, color='g')

# Plot the probability density function (PDF) of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = x*norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

title = "Fit results: mu = %.2f,  std = %.2f" % (mu, sigma)
plt.title(title)
plt.xlim(0,4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)
w=np.linspace(-1,1,20)
x,w = np.meshgrid(x,w)

def exact(x,w):
  return np.exp(x*w*1j)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(x, w, exact(x,w), cmap='viridis')

# Add a color bar which maps values to colors
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set a title for the plot
ax.set_title('3D Plot Example')

plt.show()
```

```python
def expansion(x,w):
  expans=np.zeros((len(w),len(x)))
  for l in range(3):
    expans.astype("complex128")
    expans+=(2*l+1)*(1j)**l*dot(jn(l,x),P(l)(w))

  return expans

z=expansion(x,w)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(x, w, z, cmap='viridis')

# Add a color bar which maps values to colors
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set a title for the plot
ax.set_title('3D Plot Example')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate polar coordinates
theta = np.linspace(0, 2*np.pi, 100)
r = np.linspace(0, 1, 100)

theta, r = np.meshgrid(theta, r)

# Convert polar coordinates to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)
z = np.sin(r)

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(x, y, z, cmap='viridis')

# Add a color bar which maps values to colors
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set a title for the plot
ax.set_title('3D Plot with Polar Coordinates')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate spherical coordinates
phi = np.linspace(0, np.pi, 100)
theta = np.linspace(0, 2*np.pi, 100)

phi, theta = np.meshgrid(phi, theta)

# Spherical to Cartesian coordinates conversion
x = np.sin(phi) * np.cos(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(phi)

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(x, y, z, cmap='viridis')

# Add a color bar which maps values to colors
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set a title for the plot
ax.set_title('3D Plot with Spherical Coordinates')

plt.show()
```
