```python
import numpy as np

Exact=np.log(4)-3/4

def f(x):
  return x*np.log(x)

#Trapezoid
def trap(xL, xR, h):
  n=int((xR-xL)/h)
  sum=0
  for i in range(1,n):
    x=xL+i*h
    sum+=f(x)
  return (f(xL)+f(xR)+2*sum)*h/2

#Simpson
def simp(xL,xR, h):
  n=int((xR-xL)/h)
  sum_odd=0
  sum_even=0
  for i in range(1,n):
    x=xL+i*h
    if i%2==1:
      sum_odd+=f(x)
    if i%2==0:
      sum_even+=f(x)
  return (f(xL)+f(xR)+4*sum_odd+2*sum_even)*h/3
```

```python
#Table
import pandas as pd
import numpy as np
table= pd.DataFrame(
    {'Trapezoid' : [abs(Exact-trap(1,2,1/2)),abs(Exact-trap(1,2,1/4)),
                    abs(Exact-trap(1,2,1/8)),abs(Exact-trap(1,2,1/16)),abs(Exact-trap(1,2,1/32))],
     'Simpson'    : [abs(Exact-simp(1,2,1/2)),abs(Exact-simp(1,2,1/4)),
                     abs(Exact-simp(1,2,1/8)),abs(Exact-simp(1,2,1/16)),abs(Exact-simp(1,2,1/32))]},
     index= [r'h=1/2', r'h=1/4', r'h=1/8',r'h=1/16',r'h=1/32']
)
table=table.T
```
